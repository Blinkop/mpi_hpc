```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
files = ['send.csv', 'ssend.csv', 'bsend.csv', 'rsend.csv', 'isend.csv', 'sendrecv.csv']
labels = ['send', 'ssend', 'bsend', 'rsend', 'isend', 'sendrecv']

plt.figure(figsize=(16, 9))

for i, f in enumerate(files):
    data = pd.read_csv(f)
    
    x = np.arange(10) + 1
    y = data.time.values
    
    plt.plot(x, y, 'C'+str(i)+'-', label=labels[i])
    plt.xlabel('Message size in 10^5 bytes', fontsize=20)
    plt.ylabel('System clock', fontsize=20)
    plt.legend(prop={'size': 20})
```
